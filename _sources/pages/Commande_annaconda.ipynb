# Les environnements virtuels sur Annaconda

## Création d'un environnement virtuel :


Pour creer un environnement virtuel, il faut utiliser la commande suivante :

    conda create --name <envname> python=<version>
    
<div>

</div>
    
où envname dans un nom arbitraire pour votre environnement virtuel, et version! est une version spécifique de Python que vous souhaitez configurer.
    
  
## Lister les environnements :
   
La commande est la suivante : 

    conda env list
    
<div>

</div>


## Activer un environnement :

    conda activate <envname>                                    
    conda deactivate


## Supprimer un environnement :

    conda env remove -n <envname>
   
    
## Installer un package :

    conda install -n yourenvname [package] 
    
    pip install -n yourenvname [package] 
    
    
## Lister les packages :

    conda list
    
<div>

</div>
    
## YAML :

    Type de fichier utilisé pour les serialisation des données
    
### Création d'un environnement a partir d'un fichier .yml :

    conda env create --file=myfile.yml
    
### Export d'un environnement en fichier .yml :

    conda env export > environment.yml
    
## Executer un fichier python depuis Annaconda :

Depuis annaconda prompt, se placer dans le dossier ou se trouve notre fichier python puis executer la commande suivante :

    python myfile.py

</div>
